오디오 데이터를 이해하려면 최소한 아래의 개념은 이해를 해야합니다.

- 샘플링 레이트 : 1초를 몇 등분 하는가 -> 가장 중요한 개념입니다.
- bit-depth : 소리의 크기를 몇 개의 수로 나타내는가.
- 그 외 wav의 헤더 내용 : mono/stereo, 용량, 엔디안 등...  

일단 파이썬에서 자주 쓰는 오디오 도구들을 이용해서 wav 파일을 보도록 합시다.  
저는 librosa, scipy, soundfile을 주로 사용했습니다.

```python
import librosa
import scipy
import soundfile as sf
import numpy as np
```

```python
wav1, sr1 = librosa.load('echo1.wav')
wav2, sr2 = sf.read('echo1.wav')
#sr3, wav3 = scipy.io.wavfile.read('echo1.wav')
```

wav1, wav2 배열은 -1에서 1 사이의 값으로 정규화된 넘파이 배열입니다.  
원래는 -32768 ~ 32767 사이의 16비트 정수값을 가지며, scipy로 열면 확인가능합니다.  

```python
print(wav1.shape, sr1, np.max(wav1), np.min(wav1))
print(wav2.shape, sr2, np.max(wav2), np.min(wav2))
print((wav2 * 32768)[:20])
```

```python
import IPython.display as ipd

ipd.Audio(wav1, rate=sr1)
```

```python
ipd.Audio(wav2*32768, rate=sr2)
```

ipd.Audio의 인자로 wav 파일이 아닌 배열을 넣으면 샘플링 레이트를 rate로 지정해줘야 합니다.  
wav파일을 넣으면 wav파일의 헤더에 샘플링 레이트가 있으므로 그대로 재생합니다.   
원래 샘플링 레이트보다 작거나 큰 샘플링 레이트를 넣으면 이상하게 들리는 걸 확인 가능합니다.  

```python
wav1_16k = librosa.resample(wav1, orig_sr=22050, target_sr=16000)
wav1_8k = librosa.resample(wav1, orig_sr=22050, target_sr=8000)
ipd.Audio(wav1_16k, rate=16000)
#ipd.Audio(wav1_16k, rate=12000)
```

샘플링 레이트와 관련된 법칙으로 나이퀴스트 법칙이 있습니다.  
샘플링 레이트의 절반 이하 주파수인 신호는 완벽하게 복원할 수 있다는 내용입니다.  
인간의 최대 가청주파수는 최대 20000Hz 정도인데요.  
8000Hz로 샘플링했다면 4000Hz까지의 정보를 가지고 있다는 말이죠. 차이가 느껴지시나요?  

```python
ipd.Audio(wav1_8k, rate=8000)
#ipd.Audio(wav1_8k, rate=16000)
```
